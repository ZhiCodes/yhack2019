```
tweet_text = data['results'][0]['extended_tweet']['full_text']
tweet_text
```

```
tweet_user_location = data['results'][0]['user']['location']
tweet_user_location
```

```
text = []
location = []

tweets = data['results']
for tweet in tweets:
    #get full text and location
    try:
        t = tweet['extended_tweet']['full_text']
    except:
        t = tweet['text']
    l = tweet['user']['location']

    if t is not None:
        text.append(t)
    else:
        text.append('')

    if l is not None:
        location.append(l)
    else:
        location.append('')

data_dict = {'text':text, 'location':location}
df = pd.DataFrame(data_dict)
print(df.shape)
df.head(10)
```
